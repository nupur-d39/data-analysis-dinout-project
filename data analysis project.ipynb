# DINEOUT_RESTAURANTS DATA ANALYSIS PROJECT


**About Dataset**

With smaller staff, capital and costing requirements, the restaurant business is on a consistent rise. Identifying a successful restaurant demands more than customer ratings. The dataset is a record of ratings, cost and location of all Dineout restaurants in India.

**Content**

Name: Restaurant name

Location: Complete location of the restaurant. It consists of locality and city.

Locality: Locality of the restaurant within the city.

City: City of the restaurant.

Cuisine: Types of cuisines offered by the restaurant.

Rating: Rating of the restaurant given by a number of users.

Votes: Number of customers rating the restaurant.

Cost: Approximate cost for two people.

**Inspiration**

How customers attract?

Every restaurant has a different story. Success of a restaurant depend on the location, cuisines, cost or ratings. Number of votes determine the number of customer visits and rating signifies the restaurant liking. An analysis on dataset can reveal the customer behaviour.

# DATA CLEANING


```python
#importing liabraries as per requirement
#To use numpy, you first need to import it in your Python script or notebook.
import numpy as np
#To use Pandas, you first need to import it in your Python script or notebook.
import pandas as pd
# matplotlib.pyplot is used to create the plots for the data.
import matplotlib.pyplot as plt
import seaborn as sns

```

# Inference: 

The required libraries (NumPy, Pandas, Matplotlib, and Seaborn) are imported for data manipulation and visualization.

```python
# pd.read_csv imports the file and read the data frame.
df=pd.read_csv("dineout_restaurants.csv")
```

```python
#The drop_duplicates() function removes duplicate rows from a Pandas DataFrame or Series.
df=df.drop_duplicates()
```

```python
#The df.info() function provides a summary of a Pandas DataFrame.
print(df.info())
```

# Inference: 

The structure and summary of the dataset, including column names, data types, and missing values, are displayed.

The dataset contains 2997 restaurants, with complete data across all columns. No missing values were found, ensuring a clean dataset for analysis. 
The presence of categorical and numerical variables allows for diverse insights, such as cuisine preferences, rating trends, and cost variations.

```python
# df.head() print the values of first five indeces
df.head()
```

```python
# df.tail() print the values of last five indeces
df.tail()
```

```python
#str.title() capitalizes the first letter of each word.
#str.strip() removes extra spaces at the beginning and end.

df['City'] = df['City'].str.title().str.strip()
df['Cuisine'] = df['Cuisine'].str.title().str.strip()
```

# Inference:

Ensuring uniform capitalization and spacing removes inconsistencies that might affect grouping and analysis.
 
This step enhances data reliability by preventing redundant categories caused by formatting issues.

```python
duplicates = df.duplicated().sum()

print(f"\nNumber of Duplicate Entries: {duplicates}")
#The drop_duplicates() function removes duplicate rows from a Pandas DataFrame or Series.
df=df.drop_duplicates()
```

# Inference: 

No duplicate records were found, confirming that each restaurant entry is unique. 

This maintains the accuracy of the dataset, preventing inflated counts or misleading patterns.

```python
# df.dropna() is used to remove missing (NaN or None) values from a DataFrame or Series.
df.dropna()
```

```python
# inplace=true modifies or updates data directly in the database without creating a new table or temporary storage.
df.dropna(inplace=True)
df.tail()
```

# Inference:

Since no missing values were detected, data completeness is ensured. 

This eliminates the need for imputation or deletion, allowing for seamless analysis without data gaps.



# DATA ANALYSIS


```python
# Display descriptive statistics for numeric columns
summary_statistics = df.describe()

# Print the summary statistics
print("Summary Statistics:")
summary_statistics
```

# Inference: 

Summary statistics (count, mean, std, min, max, quartiles) for numerical columns are generated to understand data distribution.

The average restaurant rating is around 4.14, with most ratings falling in a positive range. The number of votes varies significantly, with some restaurants being highly reviewed while others receive little attention. 

The cost distribution suggests the presence of both affordable and high-end restaurants.

```python
#Using .unique(): Returns an array of unique values.
print(f"Unique cities count: {df['City'].unique()}")
print(f"Unique cuisines: {df['Cuisine'].unique()}")

#Using .nunique(): Returns the count of unique values.
print(f"Unique cities count: {df['City'].nunique()}")
print(f"Unique cuisines: {df['Cuisine'].nunique()}")
```

# Inference:
The dataset covers restaurants from seven major cities, with Bangalore, Delhi, and Mumbai having the highest representation.

 The presence of 1223 unique cuisines indicates a highly diverse food scene, though some cuisines may have very few occurrences.

```python
#.value_counts() function is used to count the number of occurrences of each unique value in a pandas Series
# .head() return the first five unique value from the table.
print("\n Top 5 Most Frequent Cities:")
print(df['City'].value_counts().head())
```

# Inference:

Bangalore has the highest number of restaurants, followed by Delhi and Mumbai. 

This suggests a thriving food industry in these cities. Other cities like Gurgaon and Noida have significantly fewer restaurants, indicating a more limited dining scene.

```python
# High-rated and low-rated restaurants
# FILTERING DATA
high_rated = df[df['Rating'] >= 4.5]
low_rated = df[df['Rating'] < 3.0]

print(f"\nHigh-Rated Restaurants (4.5+): {len(high_rated)}")
print(f"Low-Rated Restaurants (<3.0): {len(low_rated)}")
```

# Inference:

A majority of restaurants have ratings above 4.0, showing a trend of positive customer feedback. Very few restaurants have ratings below 3.0, suggesting that low-rated establishments are rare in this dataset.

```python
#.max(): Returns the highest vote count in the dataset.

# Most reviewed restaurant
most_reviewed = df[df['Votes'] == df['Votes'].max()]
print("\nMost Reviewed Restaurant:")
print(most_reviewed[['Name', 'City', 'Votes', 'Rating']])
```

# Inference:

The most popular restaurant has received 5016 votes, significantly higher than the average. 

This indicates strong customer engagement, likely due to brand recognition, quality service, or aggressive marketing.


```python
#.min(): Returns the least vote count in the dataset.

# Least reviewed restaurant
least_reviewed = df[df['Votes'] == df['Votes'].min()]
print("\nLeast Reviewed Restaurant:")
print(least_reviewed[['Name', 'City', 'Votes', 'Rating']])
```

# Inference:

Several restaurants have only 1 vote, suggesting they might be new, niche, or not well-known. 

Despite low votes, some of these have perfect ratings, which might indicate bias from limited reviews.

```python
#Use isnull() to identify missing values in the DataFrame   
# use .sum() to sum the True values along each column

pd.isnull(df).sum()
```

# Inference:

No missing values were found in any column, confirming data completeness. 

This ensures that analysis can be performed without any concerns about data gaps.

```python
#df.groupby('City')['Cost'].mean(): Groups the dataset by city and calculates the average cost of restaurants in each city.

#.sort_values(ascending=False): Sorts the cities in descending order based on their average restaurant cost.

# Cost distribution by city
city_cost = df.groupby('City')['Cost'].mean().sort_values(ascending=False)
print("\nAverage Cost by City:")
print(city_cost)
```

# Inference:

The cost of dining varies significantly between cities.

 Metro cities likely have higher average costs, reflecting a more expensive dining culture compared to smaller cities.

#    DATA VISUALIZATION

```python
# Rating Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Rating'], bins=20, kde=True, color='teal')
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```

#  Inference:

The rating distribution is right-skewed, with most restaurants receiving ratings between 3.5 and 4.5. Very few restaurants have extremely low ratings (below 2.5). 

The density curve suggests that a majority of restaurants tend to have good ratings, indicating overall positive customer feedback. 

```python
# Cost Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Cost'], bins=20, kde=True, color='green')
plt.title("Distribution of Cost for Two People")
plt.xlabel("Cost")
plt.ylabel("Frequency")
plt.show()
```

# Inference:

The cost distribution is right-skewed, with most restaurants having moderate pricing. 

A small number of high-end restaurants drive up the maximum cost, but they are outliers in the dataset.

```python
# Top 10 Cities with Most Restaurants
plt.figure(figsize=(10, 5))
df['City'].value_counts().head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Cities with Most Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

# Inference:

Bangalore, Delhi, and Mumbai dominate the restaurant count, highlighting them as major food hubs. Cities with fewer restaurants may have a smaller population or a less developed food service industry.

```python
# Top 10 Popular Cuisines (Pie Chart)
top_cuisines = df['Cuisine'].value_counts().head(10)
plt.figure(figsize=(10, 10))
plt.pie(top_cuisines, labels=top_cuisines.index, autopct='%1.5f%%', startangle=140)
plt.title("Top 10 Popular Cuisines")
plt.show()
```

# Inference:

North Indian cuisine appears to be the most dominant, followed by other popular categories. 

This indicates a strong customer preference for traditional Indian dishes across multiple cities.

```python
# Cost Trends by City (Line Chart)
plt.figure(figsize=(10, 5))
city_cost.plot(kind='line', marker='o', color='blue')
plt.title("Average Cost Trends by City")
plt.xlabel("City")
plt.ylabel("Average Cost")
plt.xticks(rotation=45)
plt.show()
```

# Inference:

Cost trends suggest that some cities have consistently higher restaurant prices. The variation in cost could be influenced by economic factors, customer preferences, and the presence of luxury dining establishments.
